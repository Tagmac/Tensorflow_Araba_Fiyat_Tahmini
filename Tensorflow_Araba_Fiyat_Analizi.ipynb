```
dataFrame = dataFrame.drop("transmission",axis=1) # "Transmission" değerleri numeric olmadığı için onu veri setimizden çıkarıyoruz.
```

```
y = dataFrame["price"].values # y' ye fiyatı eşitledik ve numpy dizisine çevirdik
x = dataFrame.drop("price",axis=1).values # "dataFrame" den "price" kolonunu çıkardık ve x' e eşitledik.
```

```
y # y değerini çağırdık.
```

```
x # x değerini çağırdık.
```

```
from sklearn.model_selection import train_test_split # "sklearn.model_selection" dan "train_test_split" i çağırdık. "train_test_split" bizim verirlerimizi "test" ve "train" olmak üzere 2'ye ayırmamıza yardım edecek.
```

```
x_train, x_test, y_train, y_test = train_test_split(x,y,test_size=0.30,random_state=10) # x ve y değerlerini "test" ve "train" olmak üzere 2'ye ayırdık. "Train" i modelimizi eğitmek için "Test" i ise test etmek için kullanacağız.
```

```
len(x_train) # "x_train" deki değerlerin sayısına bakıyoruz.
```

```
len(x_test) # "x_test" teki değerlerin sayısına bakıyoruz.
```

```
from sklearn.preprocessing import MinMaxScaler # "sklearn.preprocessing" den "MinMaxScaler" ı çağırdık. "MinMaxScaler" x değerlerini ölçeklendirmek için kullandığımız kod.
```

```
scaler = MinMaxScaler() # "MinMaxScaler" dan obje oluşturduk.
```

```
x_train = scaler.fit_transform(x_train) # "x_train" i fit ederek 0 ile 1 arasındaki değerlere eşitledik.
```

```
x_test = scaler.transform(x_test) 
```

```
from tensorflow.keras.models import Sequential # Modeli oluşturacağımız sınıf
from tensorflow.keras.layers import Dense # Modelde kullanacağımız katmanlar
```

```
x_train.shape
```

```
model = Sequential()

model.add(Dense(12,activation="relu")) # 12 tane nöron verdik.
model.add(Dense(12,activation="relu")) # 12 tane nöron verdik.
model.add(Dense(12,activation="relu")) # 12 tane nöron verdik.
model.add(Dense(12,activation="relu")) # 12 tane nöron verdik.
model.add(Dense(12,activation="relu")) # 12 tane nöron verdik.
                                       # Yukarıdaki nöron verme işleminden 5 tane yaptığımız için bizim 5 katmanımız var.
                                       # "Activation" olarak "relu" seçtik.
                                       
model.add(Dense(1)) # 1 tane çıkış katmanı verdik.

model.compile(optimizer="adam",loss="mse") # "adam"  daha iyi sonuç verdiği için optimezer olarak "adam" kullanıyoruz. "loss" ise bize kayıpları veriyor."loss" değeri 0' a ne kadar yakın olursa modelimiz o kadar iyi eğitilmiş demektir.
```

```
model.fit(x = x_train, y = y_train,validation_data=(x_test,y_test),batch_size=250,epochs=400) # x ve y için train verilerimizi x ve y' ye eşitledik. 
# "validation_data" ile "x_test" ve "y_test" arasındaki kıyaslama işlemini aynı ekranda görmemizi sağlıyor.
# "batch_size" verileri part part vermemizi sağlıyor. Çünkü 13000 veriyi aynı anda verirsek modeli yorarız. "batch_size=250" diyerek 1 seferde 250 veri veriyoruz.
```

```
kayıpVerisi = pd.DataFrame(model.history.history) # Çıkan sonucu "kayıpVerisi" adıyla oluşturduğumuz yeni bir değişkene tanımlıyoruz.
```

```
kayıpVerisi.head() # "kayıpVerisi" nin ilk 5 sıradaki verilerini alıyoruz."loss" ve "val_loss" değerleri ne kadar yakın olursa modelimiz o kadar iyi eğitilmiş demektir.
```

```
kayıpVerisi.plot() # "kayıpVerisi" nin grafiğini Matplotlib ile çiziyoruz.
```

```
from sklearn.metrics import mean_squared_error, mean_absolute_error # Bu hatalar matematiksel işlemler ile x ve y karşılaştırmasını bize vermektedir.
```

```
tahminDizisi = model.predict(x_test) # "x_test" verilerini ileride kullanmak üzere "tahminDizisi" adlı değişkene tanımladık.
```

```
tahminDizisi # "tahminDizisi" ni çağırdık.
```

```
mean_absolute_error(y_test,tahminDizisi) # Fiyatın sapma oranını bulduk.
```
